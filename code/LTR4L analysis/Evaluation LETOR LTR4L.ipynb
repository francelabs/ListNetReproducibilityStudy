```python
import pandas as pd
import numpy as np
import os

from evaluation_tools import *
from script_maker import *
from itertools import product
```

# Evaluating LTR4L ListNet implementation

## MQ2008 dataset

```python
### GET MODELS' DETAILS

epochs=[100,500,1000]
learning_rates=[0.001,0.075]
mon=[0.01,0.1,0.3]
out_act=['Identity','Sigmoid']
hid_act= ['Identity','Sigmoid','ReLu']
number_hidden_layers = [0,1]
number_neurons = [15,46]

if 0 in number_hidden_layers:
    total_it = len(epochs)*len(learning_rates)*len(mon)*len(out_act)
    total_it += len(epochs)*len(learning_rates)*len(mon)*len(out_act)*len(hid_act)\
    *(len(number_hidden_layers)-1)*len(number_neurons)
else:
    total_it = len(epochs)*len(learning_rates)*len(mon)*len(out_act)*len(hid_act)\
    *len(number_hidden_layers)*len(number_neurons)

columns=["epochs","learning_rate","momentum","output function", "hidden function", "nb hidden layers", "nb neurons"]
dt=pd.DataFrame(columns=columns)

#No need to use hidden activation function or number of neurons per layers when there are no hidden layers
if 0 in number_hidden_layers:
    for i in product(epochs,learning_rates,mon, out_act):
        dt=dt.append(dict(zip(columns,i)),ignore_index=True)
    number_hidden_layers.remove(0)

#When there are hidden layers
for i in product(epochs,learning_rates,mon, out_act,hid_act,number_hidden_layers,number_neurons):
    dt=dt.append(dict(zip(columns,i)),ignore_index=True)
```

```python
### COMPUTE THE EVALUATION SCORES
path_of_script='./2020_10_04_script_score_MQ2008.sh'
low_bound=1
high_bound=total_it

name='ltr4l'#name that we gave to the linear models
data_fold='MQ2008'#Folder we are going to inspect... LETOR 4
set_evaluation_script(low_bound,high_bound,Folds=[1,2,3,4,5],
                      path_of_script=path_of_script,
                      data_fold=data_fold,name=name)
os.system(path_of_script)
```

```python
### GET EVALUATION SCORES

data_fold = "MQ2008"
name = "ltr4l"

metrics_df=get_summary_metrics(low_bound,high_bound,return_test=True,
                               name=name,Folds=[1,2,3,4,5],data_fold=data_fold)
```

```python
### GET TRAINING TIME

data_fold = "MQ2008"
name = "ltr4l"
folds = [1,2,3,4,5]
   
duration_mean = get_training_duration(data_fold, name, folds)
```

```python
### JOIN THE DATA IN ONE DATAFRAME

perf = dt.join(metrics_df, how="outer")
perf["duration_mean"] = duration_mean
```

```python
perf.sort_values(['NDCG mean 1-10 vali'],ascending=False).head(10)
```

The results obtained by the algorithm are slightly better than the baselines provided by Microsoft

```python
write_metrics_to_csv(low_bound,high_bound,return_test=True,
                     name=name,Folds=[1,2,3,4,5],data_fold=data_fold,
                     model_dataframe = dt)
```
