```python
import pandas as pd
import numpy as np
import os
```

# MQ2008

```python
path = "ListNet.MQ2008.txt"
path_baseline_prefix = "../MQ2008/results_summary/Results_baseline"

baselines_file = open(path)
```

```python
line = baselines_file.readline()
while line:
    if "testing" in line:
        line = baselines_file.readline()
        break
    else:
        line = baselines_file.readline()

###Write test file
temp_path_test = "temp_test.txt"
file_write = open(temp_path_test,'w')
while line:
    if "validation" in line:
        line = baselines_file.readline()
        break
    else:
        file_write.write(line)
        line = baselines_file.readline()
file_write.close()
data_test = pd.read_csv(temp_path_test, sep='\t')
os.remove(temp_path_test) 
        
###Write vali file
temp_path_vali = "temp_vali.txt"
file_write = open(temp_path_vali,'w')
while line:
    if "training" in line:
        line = baselines_file.readline()
        break
    else:
        file_write.write(line)
        line = baselines_file.readline()
file_write.close()
data_vali = pd.read_csv(temp_path_vali, sep='\t')
os.remove(temp_path_vali) 
        
###Write train file
temp_path_train = "temp_train.txt"
file_write = open(temp_path_train,'w')
while line:
    file_write.write(line)
    line = baselines_file.readline()
file_write.close()
data_train = pd.read_csv(temp_path_train, sep='\t')
os.remove(temp_path_train) 

baselines_file.close()
```

```python
data_test
```

```python
###REARRANGE THE TEST DATAFRAME TO BE USABLE
data_test = data_test.set_index('Folds')
data_test = data_test.drop("MeanNDCG", axis=1)

NDCG = data_test.iloc[:6]
Precision = data_test.iloc[6:]
Precision = Precision.rename(columns=Precision.iloc[0].str.replace('Pre', 'P', regex=True))
Precision = Precision.drop("Folds",axis=0)

ndcg_mean = NDCG.astype(float).values.mean(axis=1)
ndcg_var = NDCG.astype(float).values.var(axis=1)
NDCG["NDCG_Mean"] = ndcg_mean
NDCG["NDCG_Var"] = ndcg_var

pre_mean = Precision.astype(float).values.mean(axis=1)
pre_var = Precision.astype(float).values.var(axis=1)
Precision["P_Mean"] = pre_mean
Precision["P_Var"] = pre_var

data_test = Precision.join(NDCG)

path_baseline_test = "../../MQ2008/results_summary/Results_baseline_test.csv"

data_test.to_csv(path_baseline_test)
```

```python
###REARRANGE THE VALIDATION DATAFRAME TO BE USABLE
data_vali = data_vali.set_index('Folds')
data_vali = data_vali.drop("MeanNDCG", axis=1)

NDCG = data_vali.iloc[:6]
Precision = data_vali.iloc[6:]
Precision = Precision.rename(columns=Precision.iloc[0].str.replace('Pre', 'P', regex=True))
Precision = Precision.drop("Folds",axis=0)

ndcg_mean = NDCG.astype(float).values.mean(axis=1)
ndcg_var = NDCG.astype(float).values.var(axis=1)
NDCG["NDCG_Mean"] = ndcg_mean
NDCG["NDCG_Var"] = ndcg_var

pre_mean = Precision.astype(float).values.mean(axis=1)
pre_var = Precision.astype(float).values.var(axis=1)
Precision["P_Mean"] = pre_mean
Precision["P_Var"] = pre_var

data_vali = Precision.join(NDCG)

path_baseline_vali = "../MQ2008/results_summary/Results_baseline_vali.csv"

data_vali.to_csv(path_baseline_vali)
```

```python
###REARRANGE THE TRAIN DATAFRAME TO BE USABLE
data_train = data_train.set_index('Folds')
data_train = data_train.drop("MeanNDCG", axis=1)

NDCG = data_train.iloc[:6]
Precision = data_train.iloc[6:]
Precision = Precision.rename(columns=Precision.iloc[0].str.replace('Pre', 'P', regex=True))
Precision = Precision.drop("Folds",axis=0)

ndcg_mean = NDCG.astype(float).values.mean(axis=1)
ndcg_var = NDCG.astype(float).values.var(axis=1)
NDCG["NDCG_Mean"] = ndcg_mean
NDCG["NDCG_Var"] = ndcg_var

pre_mean = Precision.astype(float).values.mean(axis=1)
pre_var = Precision.astype(float).values.var(axis=1)
Precision["P_Mean"] = pre_mean
Precision["P_Var"] = pre_var

data_train = Precision.join(NDCG)

path_baseline_train = "../MQ2008/results_summary/Results_baseline_train.csv"

data_train.to_csv(path_baseline_train)
```
