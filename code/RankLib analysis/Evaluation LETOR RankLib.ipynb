```python
import pandas as pd
import numpy as np
import os

from evaluation_tools import *
from script_maker import *
from itertools import product
```

# Evaluating RankLib ListNet implementation

## MQ2008 dataset

```python
### GET MODELS' DETAILS
data_fold='MQ2008'

epochs=[500,1000,1500,2000]
learning_rates=[0.001,0.00001] #there are no other options for ranklib
metrics = ['NDCG@5','NDCG@10','P@5','P@10','ERR@10']


total_it=len(epochs)*len(learning_rates)*len(metrics)

columns=["epochs","learning_rate","Opti_method"]
dt=pd.DataFrame(columns=columns)

for i in (product(epochs,learning_rates, metrics)):
    dt=dt.append(dict(zip(columns,i)),ignore_index=True)
```

```python
### COMPUTE THE EVALUATION SCORES
path_of_script='./script_score_MQ2008.sh'
low_bound=1
high_bound=total_it

name='RankLib'#name that we gave to the models
data_fold='MQ2008'#Folder we are going to inspect... 
set_evaluation_script(low_bound,high_bound,Folds=[1,2,3,4,5],
                      path_of_script=path_of_script,
                      data_fold=data_fold,name=name)
os.system(path_of_script)
```

```python
### GET EVALUATION SCORES

data_fold = "MQ2008"
name = "RankLib"

metrics_df=get_summary_metrics(low_bound,high_bound,return_test=True,
                               name=name,Folds=[1,2,3,4,5],data_fold=data_fold)
```

```python
### GET TRAINING TIME

data_fold = "MQ2008"
name = "RankLib"
folds = [1,2,3,4,5]
   
duration_mean = get_training_duration(data_fold, name, folds)
```

```python
### JOIN THE DATA IN ONE DATAFRAME

perf = dt.join(metrics_df, how="outer")
perf["duration_mean"] = duration_mean
```

```python
perf.sort_values(['NDCG mean 1-10 vali'],ascending=False).head(10)
```

```python
write_metrics_to_csv(low_bound,high_bound,return_test=True,
                     name=name,Folds=[1,2,3,4,5],data_fold=data_fold,
                     model_dataframe = dt)
```
